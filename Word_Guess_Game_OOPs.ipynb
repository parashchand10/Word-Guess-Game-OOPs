#**Word Guess Game - OOPs**

A Word Guess Game built using **Python** and **Object-Oriented Programming** concepts.

This project demonstrates the practical implementation of **inheritance**, **class structure**, **loops**, **conditional logic**, and **input validation** in a real-world mini game.

```python
import random # Pick random elemets
```

```python
class Word_Guess:  # Parent Class
  def __init__(self, name):
    self.name = name
    print(f"Player name is: {self.name}")

  def levels(self):
    self.levels = {1 :["cat", "dog", "ball"],
                     2 : ["account", "loveable", "orange"],
                     3 : ["misconception", "polymorphism"] } # dict collection of keys and Pairs Values
```

```python
class Play_Game(Word_Guess):    # Child Class
  def __init__(self, name):
    super().__init__(name)
    self.steps = 0              # Calcuate number of chances


  def level_choice(self):

    super().levels()            # access parent class properties

    choice = int(input("Enter your choice (1, 2, 3): "))          # 1 - Easy, 2 - Moderate, 3 - Difficult

    word = random.choice(self.levels[choice])            # Generate random words using random()

    attempt = len(word)    # len() describe length of characters
    steps = 0

    firstw = [word[0]] + ["_"] * (len(word)-1)

    print("**Game Start**")

    while attempt > 0 and "_" in firstw:          # Loop use for until game win
      print("\nword: ", " ".join(firstw))

      print("-" * 20)
      inputn = input("Enter the letter: ")

      found = False
      for i in range(1, len(word)):
        if inputn == word[i]:
          firstw[i] = inputn
          found = True

      steps = steps + 1     # if guess letter is correct step increment

      if found:
        print("Correct Guess")
      else:
        attempt = attempt - 1   # if guess is wrong attempt decrement
        print("!Wrong Guess")
        print("Your remaining attempts are: ", attempt)


      if len(inputn) != 1 or not inputn.isalpha():    # Valid single input first
        print("Please enter a single alphabet only.")
        continue


    if "_" not in firstw:
      print("\n Final Word:", " ".join(firstw))
      print(f"\nCongratulation You Win and Steps are: {steps}")
    else:
      print("Game exit")
```

```python
guess = Play_Game("ray")
```

```python
guess.level_choice()
```
